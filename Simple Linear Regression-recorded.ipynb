```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('height-weight.csv')
df
```

```python
plt.scatter(df.Weight, df.Height)
plt.xlabel("weight")
plt.ylabel("Height")
plt.show()
```

```python
# problem statement- i want to predict height based on weight
```

```python
X = df[['Weight']]
y = df['Height']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# scalling(optional)-- standardization
```

```python
# fit- learning mu and sigma
# transformation- apply mu and sigma , calculation 
# if apply fit on test data then data leakage happens of test data to train data
```

```python
# so we use fit_transform in train and transform in test data 

# most of the ml model use standardscaler is used as it deals with nomrmal distribution
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
# scaling should be done always after train test split
# target variable should not be scaled
# avoid scaling categorical features
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
# scaling is optional as it does not effect the distribution of data
plt.scatter(X_train, y_train)
```

```python
# model training
from sklearn.linear_model import LinearRegression
```

```python
# linear regression parameters
# fit_intercept -- the best fit line will have intercept, by default it is true
# rest parametrs written in notes copy
```

```python
regressor  = LinearRegression()
```

```python
regressor 
```

```python
regressor.fit(X_train, y_train)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
print("The slope and coeff of model is", regressor.coef_)
print("the intercept of model is", regressor.intercept_)
```

```python
# predict on train data
# interpretation of coef -- one unit increase in weight on an average the height increased by 17.7307072
# predicted height(y) on X_train = 160 + 17.73 * (X_train)
# predicted height(y) on X_test = 160 + 17.73 * (X_test)
```

```python
# prediction on train data
regressor.predict(X_train)
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regressor.predict(X_train), color='r')
```

```python
y_pred_test = regressor.predict(X_test)
```

```python
y_pred_test
```

```python
y_pred_test, y_test
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test,regressor.predict(X_test), color='r') 
```

```python
# performance metrics
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred_test)
```

```python
mean_squared_error(y_test, y_pred_test)
```

```python
r_square = r2_score(y_test, y_pred_test) # r2 = 1-SSR/SST
r_square
```

```python
# adjusted rsquare
# adjusted R2 = [(1-rsquare)*n-1/(n-k-1)]
# n is no of observations, k is no of predictor variable
```

```python
X_test.shape
```

```python
X_test.shape[0]
```

```python
X_test.shape[1] # no of predictors
```

```python
1-(1-r_square) *(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
# adjusted rsquare will be lesser than r square
```

```python
# assumptions
# x, y have linear relationship
# observation should have no relation
# error should be constant variation
# error should be normally distributed
# features among themself should be least related 
```

```python
plt.scatter(y_test, y_pred_test)
# to check if y_test and its prediction aligns
```

```python
# error should be normally distributed
error = y_test-y_pred_test
error
```

```python
import seaborn as sns
sns.distplot(error) # approximately normal distribution
```
